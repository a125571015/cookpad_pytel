```python
import requests
import os
import re
import pandas as pd
from bs4 import BeautifulSoup


url="http://www.cpbl.com.tw/stats/all.html?&game_type=01&online=0&year=2020&stat=ppit"
headers=headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0","Host":"www.cpbl.com.tw"}

ss=requests.session()



res=ss.get(url,headers=headers)



cookies=ss.cookies.get_dict()
#print(ss.cookies.get_dict())

res=ss.get(url,headers=headers,cookies=cookies)

soup=BeautifulSoup(res.text,'html.parser')
#print(soup)

save_data=[]
save_tmp=[]
title_list=[]
table=soup.select('table.std_tb.mix_x')[0].select('tr')
test=soup.select('table.std_tb.mix_x')[0].select('tr')[0]
title=soup.select('table.std_tb.mix_x')[0].select('tr')[0].select('th')
td_data=soup.select('table.std_tb.mix_x')[0].select('tr')[0].select('td')


for j in title:
    save_title=j.text
    title_list.append(save_title)


#print(title_list)
save_data.append(title_list)
#print(save_data)
#print(table)
#print(test)
for i in table:
    td_list=i.select('td')
    
   
    for x in td_list:
        data=x.text
        save_tmp.append(data)
        
        #print(save_tmp)
        #print(save_data)
        
        
    save_data.append(save_tmp)    
    save_tmp=[]#重製save_tmp
print(save_data)
```

```python
import pandas as pd
from pandas.core.frame import DataFrame

data=DataFrame(save_data)#这时候是以行为标准写入的
print(data)
data.to_csv('中華職棒投手數據.csv',header=False,index=False,encoding='utf-8-sig')
```
